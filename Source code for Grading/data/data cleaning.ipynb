```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('cleaned_all_phones.csv') 

df['announcement_date'] = pd.to_datetime(df['announcement_date'])
if 'resolution' in df.columns:
# Split 'resolution' into two separate columns 'resolution_width' and 'resolution_height'
    df['resolution_width'] = df['resolution'].apply(lambda x: int(x.split('x')[0]))
    df['resolution_height'] = df['resolution'].apply(lambda x: int(x.split('x')[1]))

# Drop the original 'resolution' column
df.drop('resolution', axis=1, inplace=True)

# Display the first few rows of the dataset after transformations
df.head()
```

```python
numerical_cols = ['inches', 'battery', 'ram(GB)', 'weight(g)', 'storage(GB)', 'price(USD)', 'resolution_width', 'resolution_height']

# Plot boxplots for each numerical column to inspect for outliers
plt.figure(figsize=(12, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(4, 2, i)
    sns.boxplot(x=df[col])
    plt.title(col)
    plt.tight_layout()

plt.show()
```

```python
duplicate_rows = df[df.duplicated()]
num_duplicates = duplicate_rows.shape[0]

print(num_duplicates)
```
